```python
from dwave.system import LeapHybridSampler
from HP_Lattice_2D import *
```

```python
model = Lattice_HP_QUBO(dim = [10, 10], sequence="PPHHHHPPHPPHPHPPHHPPHPHPHPPHHH")
# sequence = from_str("HPPHPPHPH PPHPHPPH PPHHH")
# sequence = from_str("PPHHHHPPHPPHPHPPHHPPHPHPHPPHHH")
# sequence = from_str("PPHPPHHPPHHPPPPPHHHHHHHHHHPPPPPPHHPPHHPPHPPHHHHH")
# sequence = from_str("HHHHHHHHHHHH PHPH PPHH PPHH PPHP PHHP PHHP PHPP HHPP HHPP HPHP HHHH HHHH HHHH")
```

```python
sampler = LeapHybridSampler()
mysamples = [] 
for i in range(0, 5):
    print("Asking hybrid sampler for a sample")
    sampleset = sampler.sample_qubo(model.interaction_matrix(), time_limit=6) # recommended 20 for S48 and 100 for S64
    print(sampleset.info)
    print("Received sample set")
    sample0 = sampleset.record[0]
    model.print_energies(sample0.sample)
    mysamples.append(sample0.sample)
```
